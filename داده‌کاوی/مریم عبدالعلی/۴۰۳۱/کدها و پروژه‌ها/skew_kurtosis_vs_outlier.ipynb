```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis
from sklearn.linear_model import LinearRegression, HuberRegressor
from sklearn.metrics import mean_squared_error

def add_outliers(x, y, n_outliers=4, x_outlier_range=(-10, 10), y_outlier_range=(-25, 25)):
    x_outliers = np.concatenate([x, np.random.uniform(*x_outlier_range, n_outliers).reshape(-1, 1)])
    y_outliers = np.concatenate([y, np.random.uniform(*y_outlier_range, n_outliers)])
    return x_outliers, y_outliers

#Generate linear data (y = 2x + noise)
np.random.seed(0)
x = np.random.normal(0, 1, 200).reshape(-1, 1)
y = 2 * x.flatten() + np.random.normal(0, 0.5, 200)

n_outliers = 20  
x_outliers, y_outliers = add_outliers(x, y, n_outliers)

base_skewness_x = skew(x.flatten())
disturbed_skewness_x = skew(x_outliers.flatten())
base_kurtosis_x = kurtosis(x.flatten())
disturbed_kurtosis_x = kurtosis(x_outliers.flatten())

base_skewness_y = skew(y)
disturbed_skewness_y = skew(y_outliers)
base_kurtosis_y = kurtosis(y)
disturbed_kurtosis_y = kurtosis(y_outliers)

linear_reg = LinearRegression().fit(x_outliers, y_outliers)
huber_reg = HuberRegressor().fit(x_outliers, y_outliers)

y_pred_linear = linear_reg.predict(x_outliers)
y_pred_huber = huber_reg.predict(x_outliers)



plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
sns.histplot(x.flatten(), bins=20, kde=True, color='skyblue', edgecolor='black')
plt.title(f"Original X Data\nSkewness: {base_skewness_x:.2f}, Kurtosis: {base_kurtosis_x:.2f}")
plt.xlabel("X Value")
plt.ylabel("Frequency")

plt.subplot(1, 2, 2)
sns.histplot(x_outliers.flatten(), bins=20, kde=True, color='salmon', edgecolor='black')
plt.title(f"Disturbed X Data with Outliers\nSkewness: {disturbed_skewness_x:.2f}, Kurtosis: {disturbed_kurtosis_x:.2f}")
plt.xlabel("X Value")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(x_outliers, y_outliers, color="gray", label="Data with Outliers")
plt.plot(x_outliers, y_pred_linear, color="blue", label="Linear Regression")
plt.plot(x_outliers, y_pred_huber, color="red", linestyle="--", label="Huber Regression")
plt.plot(x_outliers,2*x_outliers.flatten(), color="cyan", linestyle="-", label="True model")
plt.title("Linear Regression vs Huber Regression on Data with Outliers")
plt.xlabel("X Value")
plt.ylabel("Y Value")
plt.legend()
plt.show()

```
